```python
img_counts = defaultdict(int) 
for path in glob.glob("Data/**/*.png",recursive=True):
    directory = os.path.dirname(path)
    name = os.path.dirname(os.path.dirname(directory[5:]))
    img_counts[name] += 1
    

img_counts
```

#### image size distribution

```python
img_sizes = defaultdict(int)
for path in glob.glob("Data/**/*.png",recursive=True):
    width, height = imagesize.get(path)
    if (width, height) not in img_sizes:
        print(path, "has this size ", (width, height))
    img_sizes[(width, height)] +=1
img_sizes
```

```python
plt.rcParams["figure.figsize"] = (10,10)

fig, ax = plt.subplots()
ax.bar([str(x) for x in img_sizes.keys()], img_sizes.values() )
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

for c in ax.containers:
    ax.bar_label(c,[str(x) for x in img_sizes.values()])
plt.show()
```
